```python
def cleanText(text):
    text = text.strip();
    text = text.replace('$$$','')
    text = text.replace('\\ldots','…')
    text = text.replace('\\prod','∏')
    text = text.replace(u'\ufeff','')
    text = text.replace(u'\xa0', u' ')
    text = text.replace('\leq', '≤');
    text = text.replace('\le', '≤');
    text = text.replace('\cdot', '⋅');
    text = text.replace('\dots', '⋯');
    text = text.replace('_', '');
    text = text.replace('{', '');
    text = text.replace('}', '');
    text = text.replace('\frac', '');
    text = text.replace('\mathbb', '');
    text = text.replace('\\times', '×');
    text = text.replace('\\neq', '≠');
    text = text.replace('\\ne', '≠');
    text = text.replace('\\bmod', ' mod')
    text = text.replace('Tags:', '');
    text = text.replace('All submissions for this problem are available.', '');    
    for x in range(5):
        text = text.replace('\n\n', '\n');
        text = text.replace(',\n,', ',');
    return text
```

```python
for d in div.find_all("a", {'class':'problem-tag-small'}): 
    d.decompose()
```

```python
linksToProblems = {}
done = {}
```

```python
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
# driver = webdriver.PhantomJS(service_args=['--ignore-ssl-errors=true', '--ssl-protocol=TLSv1'])
# browser = new webdriver.Builder().withCapabilities(Capabilities.firefox().set("acceptInsecureCerts", true)).build();
browser = webdriver.Firefox(executable_path=geckodriver, firefox_options=options)
for key, value in linksToTags.items():
    linksToProblems[key] = []
    print(f'For {key}')
    for url in value:
        browser.get(url)
        
        wait = WebDriverWait(browser, 20)
        wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, "div.problem-tagbox-inner")))
        
        html = browser.page_source

        soup = BeautifulSoup(html, 'html.parser')
        
        divs = soup.find_all('div',{'class':'problem-tagbox-inner'})
        for div in divs:
            linksToProblems[key].append(div['id'])
    done[key] = True
    print(len(linksToProblems[key]))
    time.sleep(10)
        

browser.quit()
```

```python
codechefProblemStatements = {}
done = {}
base_url = 'https://www.codechef.com/problems/'
```

```python
# driver = webdriver.PhantomJS(service_args=['--ignore-ssl-errors=true', '--ssl-protocol=TLSv1'])
# browser = new webdriver.Builder().withCapabilities(Capabilities.firefox().set("acceptInsecureCerts", true)).build();
browser = webdriver.Firefox(executable_path=geckodriver, firefox_options=options)
for key, value in linksToProblems.items():
    codechefProblemStatements[key] = []
    print(f'For {key}')
    for pid in value:
        url = base_url + pid
        try:
            browser.get(url)
            wait = WebDriverWait(browser, 20)
            wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, "a.problem-tag-small")))
        
            html = browser.page_source

            soup = BeautifulSoup(html, 'html.parser')

            div = soup.find('div',{'id':'problem-statement'})
            for d in div.find_all("a", {'class':'problem-tag-small'}): 
                d.decompose()
            codechefProblemStatements[key].append(cleanText(div.get_text(separator="\n")))
            done[pid] = True
        except:
            print(pid)
            done[pid] = False
    print(len(codechefProblemStatements[key]))
    time.sleep(10)
        

browser.quit()
```
